```python
import pandas as pd
import math
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
import keras
from keras import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
x = pd.read_csv("data/entradas_breast.csv")
y = pd.read_csv('data/saidas_breast.csv')
```

```python
def Network(activation, optimizer, loss, units, kernel_initializer, metrics=['binary_accuracy']):

    sequential = Sequential()

    sequential.add(Dense(units=units, 
                         kernel_initializer=kernel_initializer, 
                         activation=activation, 
                         input_dim=30))
    sequential.add(Dropout(0.2))
    sequential.add(Dense(units=units,
                        kernel_initializer=kernel_initializer, 
                        activation=activation))

    sequential.add(Dense(units = 1, activation='sigmoid'))

    sequential.compile(optimizer=optimizer, loss=loss, metrics=metrics)
    
    return sequential   
```

```python
classificador = KerasClassifier(build_fn=Network)
```

```python
params = {
    "activation" : ['relu', "tanh"],
    "optimizer" : ['adam', 'adamw'],
    "loss" : ['binary_crossentropy', 'hinge'],
    'epochs' : [10, 100, 1000],
    "batch_size" : [5, 10, 15, 20],
    "units" : [4, 8, 16, 32],
    "kernel_initializer" : ['normal', 'random_uniform']
}
```

```python
grid = GridSearchCV(
    estimator=classificador,
    param_grid=params,
    scoring='accuracy',
    cv=5
)
```

```python
grid.fit(x, y)
best_params = grid.best_params_
best_score = grid.best_score_
```

```python
import pandas as pd
import math
from sklearn.model_selection import train_test_split, cross_val_score
import keras
from keras import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
x = pd.read_csv("data/entradas_breast.csv")
y = pd.read_csv('data/saidas_breast.csv')
```

```python
def Network():
    optimizer = keras.optimizers.Adam(
    learning_rate=0.001,
    decay=0.0001,
    clipvalue=0.5
    )

    sequential = Sequential()

    sequential.add(Dense(units=16, activation='relu', input_dim=30))
    sequential.add(Dense(units=16, activation='relu'))
    sequential.add(Dense(units = 1, activation='sigmoid'))

    sequential.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['binary_accuracy'])
    
    return sequential
```

```python
classificador = KerasClassifier(build_fn=Network, epochs=100, batch_size=10)
```

```python
resultados = cross_val_score(estimator=classificador, X=x, y=y, cv=10, scoring='accuracy')
```

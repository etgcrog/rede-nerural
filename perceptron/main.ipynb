# Perceptron Simples para Regressao Celsius -> Fahrenheit
## T(f) = T(c) * (9/5 + 32)

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use("dark_background")
```

```python
tf.__version__
```

```python
df_temperature = pd.read_csv("data/Celsius-to-Fahrenheit.csv")
```

```python
sns.scatterplot(df_temperature['Celsius'], df_temperature['Fahrenheit']);
```

```python
x_train = df_temperature['Celsius']
y_train = df_temperature['Fahrenheit']
```

```python
x_test = np.array(x_train.iloc[:9])
y_test = np.array(y_train.iloc[:9])
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units=1, input_shape = [1]))
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.5), loss='mean_squared_error')
```

```python
epochs_hist = model.fit(x_train, y_train, epochs=500)
```

```python
pred = model.predict(x_test)
```

```python
model.evaluate(pred, y_test)
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use("dark_background")
```

```python
bikes = pd.read_csv("data/bike-sharing-daily.csv")
```

```python
bikes.head()
```

```python
#Verificando registros nulos
plt.figure(figsize=(5,3))
sns.heatmap(bikes.isnull(), yticklabels=False, cbar=False, cmap="viridis");
```

```python
#apagando colunas nao relevantes
bikes.drop(['instant', 'registered'], axis=1, inplace=True)
```

```python
bikes['dteday'] = pd.to_datetime(bikes['dteday'], format='%m/%d/%Y')
bikes.index= pd.DatetimeIndex(bikes.dteday)
bikes.drop("dteday", axis=1, inplace=True)
```

```python
bikes.head(3)
```

```python
#Verificando correlacao
plt.figure(figsize=(12,10))
sns.heatmap(bikes.corr(), cmap="viridis", annot=True);
```

```python
bikes['cnt'].asfreq('W').plot(linewidth=3);
plt.title("Bikes usage per Week")
```

```python
bikes['cnt'].asfreq('M').plot(linewidth=3);
plt.title("Bikes usage per Month")
```

```python
sns.pairplot(bikes)
```
